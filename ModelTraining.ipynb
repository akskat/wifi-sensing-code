```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow.keras.layers import (Input, Conv1D, TimeDistributed, Flatten,
                                     LSTM, Dense, Dropout, BatchNormalization,
                                     Embedding, Concatenate, Reshape)
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

# Egendefinert label smoothing-loss
def sparse_cce_with_label_smoothing(y_true, y_pred, smooth_factor=0.1):
    """
    Manuelt label smoothing. Forutsetter at y_pred har shape (batch, num_classes)
    med softmax-aktivering (from_logits=False).
    """
    # Fjern eventuell ekstra dimensjon (batch,1) -> (batch,)
    if len(y_true.shape) > 1:
        y_true = tf.squeeze(y_true, axis=-1)

    # Antall klasser
    num_classes = tf.shape(y_pred)[-1]
    
    # Konverter sparse y_true til one-hot
    y_true_onehot = tf.one_hot(tf.cast(y_true, tf.int32), depth=num_classes)
    
    # Påfør label smoothing
    smooth_positives = 1.0 - smooth_factor
    smooth_negatives = smooth_factor / tf.cast(num_classes, tf.float32)
    y_true_smooth = y_true_onehot * smooth_positives + smooth_negatives
    
    # Vanlig categorical crossentropy
    return tf.keras.losses.categorical_crossentropy(
        y_true_smooth, y_pred, from_logits=False
    )

def movement_label_func(code, label_mapping):
    return label_mapping.get(code, f"?({code})")

print("Biblioteker importert og hjelpefunksjoner definert.")
```

```python
file_path = "CSV_Library/baseline.csv"
data = pd.read_csv(file_path)
print("Original data shape:", data.shape)

# Sjekk nødvendig kolonner
if "time_since_start" not in data.columns:
    raise ValueError("'time_since_start' mangler i datasettet.")
if "movement" not in data.columns:
    raise ValueError("'movement' mangler i datasettet.")

# Label-encode 'movement'
data['movement'] = data['movement'].astype('category')
label_mapping = dict(enumerate(data['movement'].cat.categories))
data['movement'] = data['movement'].cat.codes
print("Label mapping:", label_mapping)

# Evt. konverter 'receiver_pc' og 'sender'
if data['receiver_pc'].dtype == object:
    data['receiver_pc'] = data['receiver_pc'].astype('category').cat.codes
if data['sender'].dtype == object:
    data['sender'] = data['sender'].astype('category').cat.codes

# Definer subcarrier-kolonner
amplitude_cols  = [c for c in data.columns if c.endswith('_amplitude')]
phase_cols      = [c for c in data.columns if c.endswith('_phase')]
subcarrier_cols = amplitude_cols + phase_cols
print("Antall subcarrier-kolonner:", len(subcarrier_cols))

# Definer meta
meta_cols = ["time_since_start", "time_diff"]
meta_cols = [c for c in meta_cols if c in data.columns]

df_sub = data[subcarrier_cols].fillna(0).copy()

print("Data forberedt.")
```

```python
sequence_length = 40
step_size       = 3

X_subseq_raw = []
X_meta_raw   = []
X_sender_raw = []
X_recv_raw   = []
y_labels_raw = []
ix_array     = []

for i in range(0, len(data) - sequence_length, step_size):
    seq_sub = df_sub.iloc[i:i+sequence_length].values
    last_ix = i + sequence_length - 1
    last_row = data.iloc[last_ix]

    meta_vec  = last_row[meta_cols].values if len(meta_cols) > 0 else []
    s_id      = last_row["sender"]
    r_id      = last_row["receiver_pc"]
    mov_label = last_row["movement"]

    X_subseq_raw.append(seq_sub)
    X_meta_raw.append(meta_vec)
    X_sender_raw.append(s_id)
    X_recv_raw.append(r_id)
    y_labels_raw.append(mov_label)
    ix_array.append(last_ix)

X_subseq_raw = np.array(X_subseq_raw, dtype=np.float32)
X_meta_raw   = np.array(X_meta_raw,   dtype=np.float32)
X_sender_raw = np.array(X_sender_raw, dtype=np.int32)
X_recv_raw   = np.array(X_recv_raw,   dtype=np.int32)
y_labels_raw = np.array(y_labels_raw, dtype=np.int32)
ix_array     = np.array(ix_array,     dtype=np.int32)

print(f"X_subseq_raw shape: {X_subseq_raw.shape}, y_labels_raw shape: {y_labels_raw.shape}")
```

```python
# a) TrainVal (80%) vs Test (20%)
X_sub_trainval_raw, X_sub_test_raw, \
X_meta_trainval,   X_meta_test,     \
X_sender_trainval, X_sender_test,   \
X_recv_trainval,   X_recv_test,     \
y_trainval,        y_test,          \
ix_trainval,       ix_test = train_test_split(
    X_subseq_raw, X_meta_raw, X_sender_raw, X_recv_raw, 
    y_labels_raw, ix_array,
    test_size=0.2,
    shuffle=False
)

# b) Train (60% total) vs Val (20% total) = 25% av trainval
N_trainval = len(X_sub_trainval_raw)
val_len    = int(0.25 * N_trainval)
train_len  = N_trainval - val_len

X_sub_train_raw  = X_sub_trainval_raw[:train_len]
X_sub_val_raw    = X_sub_trainval_raw[train_len:]
X_meta_train     = X_meta_trainval[:train_len]
X_meta_val       = X_meta_trainval[train_len:]
X_sender_train   = X_sender_trainval[:train_len]
X_sender_val     = X_sender_trainval[train_len:]
X_recv_train     = X_recv_trainval[:train_len]
X_recv_val       = X_recv_trainval[train_len:]
y_train          = y_trainval[:train_len]
y_val            = y_trainval[train_len:]
ix_train         = ix_trainval[:train_len]
ix_val           = ix_trainval[train_len:]

print("Train raw:", X_sub_train_raw.shape, y_train.shape)
print("Val raw:  ", X_sub_val_raw.shape,   y_val.shape)
print("Test raw: ", X_sub_test_raw.shape,  y_test.shape)
```

```python
pca = PCA(n_components=40)

N_tr, T_tr, sub_dim = X_sub_train_raw.shape
X_sub_train_2d = X_sub_train_raw.reshape(N_tr*T_tr, sub_dim)
pca.fit(X_sub_train_2d)

def apply_pca_and_reshape(X_sub_raw):
    N, T, D = X_sub_raw.shape
    X_2d = X_sub_raw.reshape(N*T, D)
    X_pca_2d = pca.transform(X_2d)
    X_pca_3d = X_pca_2d.reshape(N, T, 40)
    return X_pca_3d

X_sub_train_pca = apply_pca_and_reshape(X_sub_train_raw)
X_sub_val_pca   = apply_pca_and_reshape(X_sub_val_raw)
X_sub_test_pca  = apply_pca_and_reshape(X_sub_test_raw)

# Skalering av meta (fit bare på train)
scaler_meta = StandardScaler()
scaler_meta.fit(X_meta_train)

X_meta_train_sc = scaler_meta.transform(X_meta_train)
X_meta_val_sc   = scaler_meta.transform(X_meta_val)
X_meta_test_sc  = scaler_meta.transform(X_meta_test)

# Reshape sub -> (batch, time, feats, channel=1)
X_sub_train_pca = X_sub_train_pca[..., np.newaxis]
X_sub_val_pca   = X_sub_val_pca[..., np.newaxis]
X_sub_test_pca  = X_sub_test_pca[..., np.newaxis]

print("Etter PCA og skalering:")
print("X_sub_train_pca:", X_sub_train_pca.shape)
print("X_meta_train_sc:", X_meta_train_sc.shape)
```

```python
seq_length = X_sub_train_pca.shape[1]  # 40
pca_feats  = X_sub_train_pca.shape[2]  # 40
meta_dim   = X_meta_train_sc.shape[1]

num_sender_unique = int(np.max(X_sender_trainval)+1)
num_recv_unique   = int(np.max(X_recv_trainval)+1)

# Inndata
sub_input    = Input(shape=(seq_length, pca_feats, 1), name="sub_input")
sender_input = Input(shape=(), dtype='int32', name='sender_input')
recv_input   = Input(shape=(), dtype='int32', name='receiver_input')
meta_input   = Input(shape=(meta_dim,), name="meta_input")

# Embeddings
emb_dim = 8
sender_emb = Embedding(num_sender_unique, emb_dim)(sender_input)
recv_emb   = Embedding(num_recv_unique,   emb_dim)(recv_input)
sender_emb = Reshape((emb_dim,))(sender_emb)
recv_emb   = Reshape((emb_dim,))(recv_emb)

emb_merged = Concatenate()([sender_emb, recv_emb])
emb_merged = Dense(16, activation='relu')(emb_merged)
emb_merged = Dropout(0.2)(emb_merged)

# Meta
m = Dense(16, activation='relu')(meta_input)
m = Dropout(0.2)(m)
meta_out = Dense(16, activation='relu')(m)

# CNN -> 2x LSTM
x = TimeDistributed(Conv1D(64, 3, padding='same', activation='relu'))(sub_input)
x = TimeDistributed(BatchNormalization())(x)
x = TimeDistributed(Dropout(0.2))(x)

x = TimeDistributed(Conv1D(128, 3, padding='same', activation='relu'))(x)
x = TimeDistributed(BatchNormalization())(x)
x = TimeDistributed(Dropout(0.2))(x)

x = TimeDistributed(Flatten())(x)

x = LSTM(128, return_sequences=True)(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

x = LSTM(64, return_sequences=False)(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

csi_out = Dense(64, activation='relu')(x)
csi_out = Dropout(0.2)(csi_out)

fusion = Concatenate()([csi_out, emb_merged, meta_out])
z = Dense(64, activation='relu')(fusion)
z = Dropout(0.3)(z)

num_classes = len(np.unique(y_train))
final_out = Dense(num_classes, activation='softmax')(z)

model = Model(
    inputs=[sub_input, sender_input, recv_input, meta_input],
    outputs=final_out
)

# Custom label smoothing (hvis TF < 2.2)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss=lambda yt, yp: sparse_cce_with_label_smoothing(yt, yp, smooth_factor=0.1),
    metrics=['accuracy']
)

model.summary()
print("Modell bygget og kompilert.")
```

```python
unique_classes = np.unique(y_train)
class_weights_array = compute_class_weight(
    class_weight='balanced',
    classes=unique_classes,
    y=y_train
)
class_weights = {cls: wt for cls, wt in zip(unique_classes, class_weights_array)}
print("Klassevekter:", class_weights)

early_stopping = EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)
reduce_lr      = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-7)
ckpt           = ModelCheckpoint("multi_input_model.h5", monitor='val_loss', save_best_only=True)

history = model.fit(
    [X_sub_train_pca, X_sender_train, X_recv_train, X_meta_train_sc],
    y_train,
    validation_data=(
        [X_sub_val_pca, X_sender_val, X_recv_val, X_meta_val_sc],
        y_val
    ),
    epochs=30,
    batch_size=64,
    callbacks=[early_stopping, reduce_lr, ckpt],
    class_weight=class_weights,
    verbose=1
)

model.load_weights("multi_input_model.h5")
print("Trening ferdig, beste vekter lastet inn.")
```

```python
# Evaluate
test_loss, test_acc = model.evaluate(
    [X_sub_test_pca, X_sender_test, X_recv_test, X_meta_test_sc],
    y_test,
    verbose=1
)
print(f"\nFinal Test Accuracy (best weights): {test_acc:.4f}")

# Lag test CSV (tider + movement)
test_data = data.iloc[ix_test].copy()
test_data.sort_values('time_since_start', inplace=True)
test_data['movement_label'] = test_data['movement'].apply(lambda c: label_mapping.get(c, f"?({c})"))

test_data_small = test_data[['time_since_start','movement','movement_label']]
test_csv = "test_times_movement.csv"
test_data_small.to_csv(test_csv, index=False)
print(f"\nTest CSV lagret: {test_csv}")
```

```python
# Kjør prediksjon (kan spesifisere batch_size for å unngå minnefeil)
y_pred_probs = model.predict(
    [X_sub_test_pca, X_sender_test, X_recv_test, X_meta_test_sc],
    batch_size=64
)
y_pred = np.argmax(y_pred_probs, axis=1)

true_str = [label_mapping.get(c, f"?({c})") for c in y_test]
pred_str = [label_mapping.get(c, f"?({c})") for c in y_pred]

df_compare = pd.DataFrame({'true': true_str, 'predicted': pred_str})
df_compare.to_csv("test_predictions.csv", index=False)
print("test_predictions.csv lagret med [true, predicted].")

labels_in_test = sorted(list(set(true_str + pred_str)))
cm = confusion_matrix(true_str, pred_str, labels=labels_in_test)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels_in_test,
            yticklabels=labels_in_test)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (Test)")
plt.show()

print("\nClassification report:")
print(classification_report(true_str, pred_str, labels=labels_in_test))
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd
import joblib  # for å laste pca, scaler

###############################################################################
# 1) Oppsett
###############################################################################

sequence_length = 40
step_size = 3
CHUNK_SIZE = 50_000  # antall rader (i-løkker) vi håndterer i slengen

# Laster inn pca og scaler_meta som du tidligere fit på train-data
pca = joblib.load("pca.pkl")          # her antar vi at du har lagret pca med f.eks. joblib.dump(pca, "pca.pkl")
scaler_meta = joblib.load("scaler.pkl")

# Laster inn csv
file_path = "CSV_Library/with_obstacles.csv"
df = pd.read_csv(file_path)
print("Leser with_obstacles.csv, shape:", df.shape)

# Oppdater kolonner
df['movement'] = df['movement'].astype('category').cat.codes
if df['receiver_pc'].dtype == object:
    df['receiver_pc'] = df['receiver_pc'].astype('category').cat.codes
if df['sender'].dtype == object:
    df['sender'] = df['sender'].astype('category').cat.codes

amplitude_cols = [c for c in df.columns if c.endswith('_amplitude')]
phase_cols     = [c for c in df.columns if c.endswith('_phase')]
subcarrier_cols = amplitude_cols + phase_cols

meta_cols = ["time_since_start","time_diff"]
meta_cols = [c for c in meta_cols if c in df.columns]

df_sub = df[subcarrier_cols].fillna(0).copy()

num_rows = len(df)
max_i = num_rows - sequence_length
print(f"max_i = {max_i}  (step={step_size}, seq_len={sequence_length})")

###############################################################################
# 2) Generator-funksjon for å hente en chunk av rader
###############################################################################
def process_chunk(start_i, end_i, chunk_index):
    """
    Bygger sekvenser for range(start_i, end_i, step=step_size).
    Kjører PCA-transform, scaler_meta.transform, 
    lagrer data i en .npz-fil (f.eks. chunk_{chunk_index}.npz).
    """
    X_subseq_list = []
    X_meta_list   = []
    X_sender_list = []
    X_recv_list   = []
    Y_list        = []
    
    # Bygg sekvenser
    for i in range(start_i, end_i, step_size):
        last_ix = i + sequence_length - 1
        if last_ix >= len(df):
            break

        seq_sub = df_sub.iloc[i : i+sequence_length].values
        last_row = df.iloc[last_ix]

        meta_vec  = last_row[meta_cols].values if len(meta_cols) > 0 else []
        s_id      = last_row["sender"]
        r_id      = last_row["receiver_pc"]
        mov_label = last_row["movement"]

        X_subseq_list.append(seq_sub)
        X_meta_list.append(meta_vec)
        X_sender_list.append(s_id)
        X_recv_list.append(r_id)
        Y_list.append(mov_label)
    
    if not X_subseq_list:
        print(f"Ingen sekvenser i chunk {chunk_index}")
        return
    
    # Konverter til NumPy
    X_subseq_raw = np.array(X_subseq_list, dtype=np.float32)
    X_meta_raw   = np.array(X_meta_list,  dtype=np.float32)
    X_sender_raw = np.array(X_sender_list, dtype=np.int32)
    X_recv_raw   = np.array(X_recv_list,   dtype=np.int32)
    y_raw        = np.array(Y_list,        dtype=np.int32)

    # PCA transform
    N_chunk, T_chunk, D_chunk = X_subseq_raw.shape
    X_2d = X_subseq_raw.reshape(N_chunk*T_chunk, D_chunk)
    X_pca_2d = pca.transform(X_2d)  # NB: pca allerede fit på train
    X_pca_3d = X_pca_2d.reshape(N_chunk, T_chunk, pca.n_components_)

    # Skalere meta
    X_meta_sc = scaler_meta.transform(X_meta_raw)

    # Reshape sub (batch,time,feats,1)
    X_pca_3d = X_pca_3d[..., np.newaxis]

    # Lagre i fil (f.eks. npz)
    out_filename = f"with_obs_chunk_{chunk_index}.npz"
    np.savez_compressed(
        out_filename,
        X_sub=X_pca_3d,
        X_meta=X_meta_sc,
        X_sender=X_sender_raw,
        X_recv=X_recv_raw,
        y=y_raw
    )
    print(f"LAGRET {out_filename}: {X_pca_3d.shape} sekvenser")

###############################################################################
# 3) Hoved-løkke som kjører prosess chunk for chunk
###############################################################################
chunk_idx = 0
i_global  = 0
while i_global < max_i:
    chunk_start = i_global
    chunk_end   = min(i_global + CHUNK_SIZE, max_i)
    
    process_chunk(chunk_start, chunk_end, chunk_idx)
    
    i_global = chunk_end
    chunk_idx += 1

print("Ferdig med all preprocessing og lagring av NPZ-chunks.")
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

import numpy as np
import pandas as pd
import tensorflow as tf
import glob
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report

from tensorflow.keras.models import load_model

model = load_model("multi_input_model.h5", compile=False)
# Deretter kan du compilen om du vil ha metrics:
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 2) Finn alle .npz-chunks
chunk_files = sorted(glob.glob("with_obs_chunk_*.npz"))
print("Fant npz-chunks:", chunk_files)

all_true_labels = []
all_pred_labels = []

# For label-strenger (movement) bruker du f.eks. en label_mapping definert manuelt:
label_mapping = {0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'O'}

# 3) For hver chunk, last og predict
for chunk_file in chunk_files:
    data_chunk = np.load(chunk_file)
    
    X_sub = data_chunk["X_sub"]       # (N_chunk, seq_length, pca_components, 1)
    X_meta = data_chunk["X_meta"]     # (N_chunk, meta_dim)
    X_sender = data_chunk["X_sender"] # (N_chunk,)
    X_recv   = data_chunk["X_recv"]   # (N_chunk,)
    y_true   = data_chunk["y"]        # (N_chunk,)
    
    # Kjør prediksjon
    y_pred_probs = model.predict(
        [X_sub, X_sender, X_recv, X_meta],
        batch_size=64
    )
    y_pred_int = np.argmax(y_pred_probs, axis=1)
    
    # Konverter til str
    true_str = [label_mapping.get(c, f"?({c})") for c in y_true]
    pred_str = [label_mapping.get(p, f"?({p})") for p in y_pred_int]
    
    all_true_labels.extend(true_str)
    all_pred_labels.extend(pred_str)

print(f"\nTotalt antall eksempler: {len(all_true_labels)}")

# 4) Confusion matrix
labels_all = sorted(list(set(all_true_labels + all_pred_labels)))
cm = confusion_matrix(all_true_labels, all_pred_labels, labels=labels_all)

import pandas as pd
df_compare = pd.DataFrame({"true":all_true_labels, "predicted":all_pred_labels})
df_compare.to_csv("with_obstacles_predictions.csv", index=False)
print("with_obstacles_predictions.csv lagret.")

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=labels_all, yticklabels=labels_all)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix (with_obstacles)")
plt.show()

print("\nClassification report:")
print(classification_report(all_true_labels, all_pred_labels, labels=labels_all))
```
